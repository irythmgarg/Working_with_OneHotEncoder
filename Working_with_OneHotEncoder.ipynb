# **WORKING WITH ONEHOTENCODER AND STANDARD SCALER**


```python
import pandas as pd# import pandas class
import numpy as np  #import numpy class
```

```python
df=pd.read_csv('/content/consumer_electronics_sales_data.csv',nrows=150); #load dataset
df.drop(columns=['ProductID'],inplace=True) # delete unwanted columns
df.head() # first five rows
```

```python
df.shape # shape of dataset
```

```python
x=df.iloc[:,0:7]; # independent columns
y=df.PurchaseIntent; #dependent column
x.head()# first five rows
```

```python
y.head()
x.shape
```

```python
x['ProductCategory'].value_counts() # count of each item of ProdectCatefory
```

```python
from sklearn.model_selection import train_test_split # import sklearn train_test_split to train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) # devide train and test in 80:20
x_train.head() # first five rows
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder # import standard scaler and One Hot Encoder
from sklearn.compose import ColumnTransformer # include column transformer
oe=OneHotEncoder(sparse=False,handle_unknown='ignore',drop='first') # create a object of onehotencoder
#sparse='False' means not sparse matric formation
#drop =first means drop first column after transformation to get rid of multicollinearity
#handle unknown =ignore means drop unknown row in test data

```

```python
ct=ColumnTransformer([
    ('tnf1',oe,['ProductCategory',	'ProductBrand'])   # apply on two columns
],remainder='passthrough') # rest columns are as such passed
```

```python
x_train_trans=pd.DataFrame(ct.fit_transform(x_train)); #. convert it into data from as conlumntransformer make it a numpy array after transformation
x_test_trans=pd.DataFrame(ct.transform(x_test));
len(x_test_trans)
```

```python
si=StandardScaler() # apply standard scaler on Price column
x_train_price=x_train_trans[[8]]
x_test_price=x_test_trans[[8]]

x_train_price=si.fit_transform(x_train_price)
x_test_price=si.transform(x_test_price)
```

```python
x_train_trans.head()
```

```python
x_train_trans[8]=x_train_price
x_test_trans[8]=x_test_price
x_train_price
```

```python
x_train_trans.head()
```

```python
x_test_trans.head()
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression # apply logistic regression algorithm as we have classification problem
lr=LogisticRegression(max_iter=1000) #create an object of logistic regression class
lr.fit(x_train_trans,y_train) # fit the training data
```

```python
y_pred=lr.predict(x_test_trans) # predict the test results
y_pred #predicted results
```

```python
y_test  #actual results
```

```python
from sklearn.metrics import accuracy_score # accuray score to check accuracy
```

```python
accuracy_score(y_test,y_pred) # calculate accuracy between real and predicted
```

```python
from sklearn.model_selection import cross_val_score # cross val score to check accuracy more than one time
```

```python
score=np.mean(cross_val_score(lr,x_train_trans,y_train,cv=10)) # means of 10 time accuray to get rid of bias results
score
```
